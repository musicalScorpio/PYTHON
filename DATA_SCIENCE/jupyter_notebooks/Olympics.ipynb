```python

#Plot the 5 athlets who have won the most gold medals over the history of Olympics

oo_mx =oo.groupby(['Athlete','Medal']).size().unstack(['Medal'],fill_value=0)
oo_mx.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']].head(5).plot(kind='bar')


```

```python
##Data Visualizations - heat map

import seaborn as sns
lo=oo[oo.Edition ==2008]
g=lo.groupby(['NOC','Medal']).size().unstack('Medal',fill_value=0)
g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]

g= g.transpose()
sns.heatmap(g)
```

```python
#Using colormaps
g=oo.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
from matplotlib.colors import ListedColormap

#Color hex
gsb = ['#dbb40c','#c5c9c7','#a87900']
my_gsb = ListedColormap(sns.color_palette(gsb))
#my_cmap = sns.palplot(sns.color_palette(gsb))

g_bplot = g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']].head()
g_bplot.plot(kind='bar', colormap=my_gsb)

```

```python
# In every Olympics which US Athlete has won the most total number of medals
#Include the Atheles discipline
gy = oo [oo.NOC == 'USA']
gy = gy.groupby(['Edition','Athlete','Medal']).size().unstack(['Medal'],fill_value=0)
gy['Total'] = gy['Gold'] +  gy['Silver'] + gy['Bronze']
gy.reset_index(inplace=True)
gy
#for year, group in gy.groupby(['Edition']):
#    print(group.sort_values('Total',ascending=False)[:1])
    
tu = [group.sort_values('Total',ascending=False)[:1] for year, group in gy.groupby(['Edition']) ]
import pandas as pd
df =pd.DataFrame()



for i in tu:
    df = pd.concat([df, pd.DataFrame(i)], ignore_index=True)

df







```
